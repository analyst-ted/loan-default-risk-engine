```python
# Load the Data

df = pd.read_csv('../data/raw/lending_club_loan_two.csv')
print('Shape:', df.shape)
df.head()
```

```python
# First Audit

print('=== BASIC INFO ===')
print(f'Rows: {df.shape[0]:,}')
print(f'Columns: {df.shape[1]}')
print(f'\n=== DATA TYPES ===')
print(df.dtypes)
```

```python
# Missing values audit

# Universal missing value scanner
missing_disguises = ['?', 'NA', 'N/A', 'na', 'none', 
                     'None', 'NONE', '-', '--', 'unknown', ' ', '']

print('=== DISGUISED MISSING VALUES ===')
for col in df.select_dtypes(include='object').columns:
    for disguise in missing_disguises:
        count = (df[col] == disguise).sum()
        if count > 0:
            print(f'{col} → "{disguise}" appears {count} times')

print('\n=== ACTUAL NULL VALUES ===')
missing = df.isnull().sum()
missing_pct = df.isnull().mean() * 100
missing_df = pd.DataFrame({
    'count': missing,
    'percentage': missing_pct
})
missing_df = missing_df[missing_df['count'] > 0].sort_values(
    by='percentage', ascending=False
)
print(missing_df)
```

```python
# Categorical Columns audit

print('=== UNIQUE VALUES PER CATEGORICAL COLUMN ===')
for col in df.select_dtypes(include='object').columns:
    print(f'{col}: {df[col].nunique()} unique values')
    print(df[col].value_counts().head(3))
    print()
```

# Complete Cleaning Strategy

| COLUMN | ACTION | REASON |
| --- | --- | --- |
|emp_title                          |Drop column                                  |173k unique values
|title                              |Drop column                                  |48k unique values
|address                            |Drop column                                  |No predictive value
|grade                              |Drop column                                  |Redundant with sub_grade
|issue_d                            |Drop column                                  |Data leakage
|earliest_cr_line                   |Drop column                                  |Raw date, needs parsing
|mort_acc                           |Impute (group mean by total_acc)             |9.5% missing, too much to drop                 
|emp_length                         |Impute with mode                             |4.6% missing
|pub_rec_bankruptcies               |Drop rows                                    |0.1% missing, tiny
|revol_util                         |Drop rows                                    |0.07% missing, tiny
|home_ownership=NONE                |Drop rows                                    |Invalid category, 31 rows
|dti outliers                       |Cap at 99th pct                              |Data entry errors
|revol_util outliers                |Cap at 99th pct                              |Mathematically impossible
|annual_inc outliers                |Cap at 99th pct                              |Extreme values hurt
|loan_status                        |Map to binary                                |Fully Paid=1, Charged Off=0

```python
# Drop irrelevant and leakage columns

cols_to_drop = [
    'emp_title',        # 173k unique values — too noisy
    'title',            # 48k unique values — redundant with purpose
    'address',          # No predictive value
    'grade',            # Redundant — sub_grade has same info + more
    'issue_d',          # Data leakage — not available at prediction time
    'earliest_cr_line'  # Raw date string — needs complex parsing
]

df = df.drop(columns=cols_to_drop)
print(f'Shape after dropping columns: {df.shape}')
print(f'Remaining columns: {df.columns.tolist()}')
```

```python
# Handle home_ownership NONE

print(f'Rows before: {len(df)}')
df = df[df['home_ownership'] != 'NONE']
print(f'Rows after removing NONE: {len(df)}')
```

```python
# Drop missing rows for pub_rec_bankruptcies and revol_util

print(f'Rows before dropping nulls: {len(df)}')
df = df.dropna(subset=['pub_rec_bankruptcies', 'revol_util'])
print(f'Rows after dropping nulls: {len(df)}')
```

```python
# Cap outliers at 99th percentile
cols_to_cap = ['dti', 'revol_util', 'annual_inc']

for col in cols_to_cap:
    cap_value = df[col].quantile(0.99)
    original_max = df[col].max()
    df[col] = df[col].clip(upper=cap_value)
    print(f'{col}: {original_max:.2f} -> capped at {cap_value:.2f}')
```
